```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
# emulate test data of 6 x 6 matrix of sequential elements
DF_obj = pd.DataFrame(np.arange(36).reshape(6, 6))
DF_obj
```

```python
# emulate test data of 5 x 3 matrix of sequential elements
DF_obj_2 = pd.DataFrame(np.arange(15).reshape(5, 3))
DF_obj_2
```

```python
# 1. concatenate two datasets
# axis - axis of concatenation (1 - by columns)
pd.concat([DF_obj, DF_obj_2], axis = 1)
```

```python
# 2. if we need to remove rows or columns:
DF_obj.drop([0, 2]) # drop 0 & 2 rows
```

```python
DF_obj.drop([0, 2], axis = 1) # drop 0 & 2 columns
```

```python
# 3. concatenate vector (Series) and matrix (DataFrame)
series_obj = Series(np.arange(6))
series_obj.name = "Vector" # just assign name for display purposes
series_obj
```

```python
concatenated = DataFrame.join(DF_obj, series_obj)
concatenated
```

```python
# 4. append a matrix to matrix

# https://stackoverflow.com/a/75956237/18619615
# appended = concatenated.append(concatenated, ignore_index = False)

# ignore_index - True: concatenated row/column indexes are replaced with new sequential ones
# ignore_index - False: keep initial indexes
appended = pd.concat([concatenated, concatenated], ignore_index=False)
appended
```

```python
appended = pd.concat([concatenated, concatenated], ignore_index=True)
appended
```

```python
# 5. sorting
# 'by' - name of the column by which we're sorting
DF_sorted = DF_obj.sort_values(by = 5, ascending = False)
DF_sorted
```
